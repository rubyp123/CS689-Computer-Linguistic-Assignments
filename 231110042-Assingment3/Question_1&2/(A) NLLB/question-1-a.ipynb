 ## Hindi to English translation

```python
from transformers import NllbTokenizer
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer
import random
```

```python
with open("/kaggle/input/dataset/test.hi", "r", encoding="utf-8") as file:
    d = file.readlines()

# # Print the contents of the file
#  data = data[0:1000]
hi_data = random.sample(d,1000)
```

```python
print(len(hi_data))

for i in range(0,len(hi_data)):
    print(hi_data[i])
```

```python
# from transformers import AutoModelForSeq2SeqLM, AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained(
    "facebook/nllb-200-distilled-600M" , src_lang="hi_Deva"
)
model = AutoModelForSeq2SeqLM.from_pretrained("facebook/nllb-200-distilled-600M")

```

```python
model=model.to("cuda")
```

```python
ans = []
for i in hi_data:
    inputs = tokenizer(i,return_tensors="pt").to("cuda")
    translated_tokens = model.generate( **inputs, forced_bos_token_id=tokenizer.lang_code_to_id["eng_Latn"], max_length=30 )
    a = tokenizer.batch_decode(translated_tokens, skip_special_tokens=True)[0]
    ans.append(a)
```

```python
for i in ans:
    print(i)
```

```python
# Open a file named "example.txt" in write mode ('w') with UTF-8 encoding
with open('Hin_to_Eng_input.txt', 'w', encoding='utf-8') as f:
    # Write some text to the file
    for i in hi_data:
        f.writelines(i+'\n')

with open('Hin_to_Eng_output.txt', 'w', encoding='utf-8') as f:
    # Write some text to the file
    for i in ans:
        f.writelines(i+'\n')
```
