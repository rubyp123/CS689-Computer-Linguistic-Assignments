```python

def remove_punctuation(sentence):
    hindi_punctuation = ['।', '॥', ',', '?', '"', "'", ';', '-', '(', ')', '[', ']', '{', '}', '...','!','.']
    for punctuation in hindi_punctuation:
        sentence = sentence.replace(punctuation, '')
    return sentence
```

```python

vowels = {
    '\u093E': 'आ',  # आ
    '\u093F': 'इ',  # इ
    '\u0940': 'ई',  # ई
    '\u0941': 'उ',  # उ
    '\u0942': 'ऊ',  # ऊ
    '\u0947': 'ए',  # ए
    '\u0948': 'ऐ',  # ऐ
    '\u094B': 'ओ',  # ओ
    '\u094C': 'औ',  # औ
     'ं' : 'अं',
     'ः' : 'अः'

}

consonants = ['क', 'ख', 'ग', 'घ', 'ङ', 'च', 'छ', 'ज', 'झ', 'ञ', 'ट', 'ठ', 'ड', 'ढ', 'ण', 'त', 'थ', 'द', 'ध', 'न', 'प', 'फ', 'ब', 'भ', 'म', 'य', 'र', 'ल', 'व', 'श', 'ष', 'स', 'ह']
v = ['अ', 'आ', 'इ', 'ई', 'उ', 'ऊ', 'ए', 'ऐ', 'ओ', 'औ']

halant = "\u094D"
aa = 'अ'

ch = 'ँ'
re = 'ऋ'
an = 'ं'


def varna(str):
    tokens=[]
    for i in  str:
        if(i==' '):
            tokens.append(" ")
            continue
        if(i>='0' and i<='9'):
            tokens.append(i)
            continue
        if(i==an):
            tokens.append('अं')
            continue
        if(i==re):
            tokens.append(re)
            continue

        if(i==ch):
            x=''
            if(len(tokens)>=1):
                x = tokens[len(tokens)-1]
                tokens = tokens[:-1]
            tokens.append(x+i)
            continue 

        if(i==halant):
            tokens = tokens[:-1]   
            continue

        for j in v:
            if(i==j):
                tokens.append(j)
                continue

        for j in consonants:
            if(j==i):
                tokens.append(j+halant)
                tokens.append(aa)

        for key in vowels:
            if(i==key):
                tokens = tokens[:-1]
                tokens.append(vowels[key])
    return tokens

# print(varna("पहले"))
```

```python
###################SYLLABLES ##########################
# inputs  = tokens

vowels_s = ['अ', 'आ', 'इ', 'ई', 'उ', 'ऊ', 'ए', 'ऐ', 'ओ', 'औ','अं']

consonants_s = [
    'क्', 'ख्', 'ग्', 'घ्', 'ङ्', 'च्', 'छ्', 'ज्', 'झ्', 'ञ्',
    'ट्', 'ठ्', 'ड्', 'ढ्', 'ण्', 'त्', 'थ्', 'द्', 'ध्', 'न्',
    'प्', 'फ्', 'ब्', 'भ्', 'म्', 'य्', 'र्', 'ल्', 'व्', 'श्',
    'ष्', 'स्', 'ह्'
]


v_to_m = {
    'अ' : '',
    'आ': '\u093E',  # आ
    'इ': '\u093F',  # इ
    'ई': '\u0940',  # ई
    'उ': '\u0941',  # उ
    'ऊ': '\u0942',  # ऊ
    'ए': '\u0947',  # ए
    'ऐ': '\u0948',  # ऐ
    'ओ': '\u094B',  # ओ
    'औ': '\u094C',  # औ
    'अं': 'ं',
    'अः': 'ः'
}


def Syllables(inputs):
    u=[]
    syllable=[]
    result = ''
    i=0
    while i < len(inputs):
        if inputs[i] in vowels_s:  
            result = result + inputs[i]
            if(i+1 < len(inputs) and inputs[i+1]=='अं'):
                result = result + inputs[i+1]
                i=i+1
            u.append(result)
            result=""

        else:
            result += inputs[i]
        i+=1



    # print(u)
    for i in u:
        str=""
        for j in i:
            if j in vowels_s:
                if len(str)==0:
                    str = str + j
                    continue
                elif(len(str)>0):
                    h = str[len(str)-1]
                    if (h==halant):
                        str = str[:-1]
                        str=str + (v_to_m[j])
                    else:
                        str = str + v_to_m[j]

                else :
                    str=str + j
            else:
                str=str+j        
        syllable.append(str) 
    return syllable
```
