```python
#################### PART 1 ###################### 
vowels = {
    '\u093E': 'आ',  # आ
    '\u093F': 'इ',  # इ
    '\u0940': 'ई',  # ई
    '\u0941': 'उ',  # उ
    '\u0942': 'ऊ',  # ऊ
    '\u0947': 'ए',  # ए
    '\u0948': 'ऐ',  # ऐ
    '\u094B': 'ओ',  # ओ
    '\u094C': 'औ',  # औ
     'ं' : 'अं',
     'ः' : 'अः'

}

consonants = ['क', 'ख', 'ग', 'घ', 'ङ', 'च', 'छ', 'ज', 'झ', 'ञ', 'ट', 'ठ', 'ड', 'ढ', 'ण', 'त', 'थ', 'द', 'ध', 'न', 'प', 'फ', 'ब', 'भ', 'म', 'य', 'र', 'ल', 'व', 'श', 'ष', 'स', 'ह']
v = ['अ', 'आ', 'इ', 'ई', 'उ', 'ऊ', 'ए', 'ऐ', 'ओ', 'औ']

halant = "\u094D"
a = 'अ'

ch = 'ँ'
re = 'ऋ'
an = 'ं'


def varna(str):
    tokens=[]
    for i in  str:
        if(i==' '):
            tokens.append(" ")
            continue
        if(i>='0' and i<='9'):
            tokens.append(i)
            continue
        if(i==an):
            tokens.append('अं')
            continue
        if(i==re):
            tokens.append(re)
            continue

        if(i==ch):
            x=''
            if(len(tokens)>=1):
                x = tokens[len(tokens)-1]
                tokens = tokens[:-1]
            tokens.append(x+i)
            continue 

        if(i==halant):
            tokens = tokens[:-1]   
            continue

        for j in v:
            if(i==j):
                tokens.append(j)
                continue

        for j in consonants:
            if(j==i):
                tokens.append(j+halant)
                tokens.append(a)

        for key in vowels:
            if(i==key):
                tokens = tokens[:-1]
                tokens.append(vowels[key])
    return tokens


ans=[]
# s = "मुख्य विपक्षी दल कांग्रेस ने संबित के बयान की आलोचना करते हुए कहा"
s = "व्याप्ति प्रत्यक्ष"
# for i in s:
    
ans = varna(s)
print(ans)
```

```python
######################## UNIGRAMS ( PART 2) ####################################
tokens=[]
with open('hi_100.txt', 'r', encoding='utf-8') as file:
    str = file.read()
    
# str = "खबर में पार्क के प्रवक्ता जैमी रिचर्ड्स के हवाले से कहा गया है 100"
# str = "100 200"

str = str.split()

# for i in str:
#     print(i)
# print(str)

for i in str:
    tokens.append(varna(i))

for i in range(0,50):
    print(tokens[i])
            
# print(tokens)
```

```python
################### UNIGRAM OF CHARACTERS ############################
uni_char={}

for i in tokens:
    for j in i:
        if j in uni_char:
            uni_char[j]+=1
        else:
            uni_char[j]=1;

sorted_uni_char = sorted(uni_char.items(), key=lambda item: item[1],reverse=True)

top_twenty = sorted_uni_char[:20]

for i,j in top_twenty:
    print(i , " : " , j)
```

```python
############################ BIGRAMS OF CHARACTERS #################################
b = []

char = [item for sublist in tokens for item in sublist]

for i in range(len(char)-1):
    b.append(char[i] + char[i+1])


# print(b)

bi_char={}

for i in b:
    if i in bi_char:
        bi_char[i]+=1
    else:
        bi_char[i]=1

sorted_bi_char = sorted(bi_char.items(), key=lambda item: item[1],reverse=True)

# print(sorted_bigrams)
top_twenty = sorted_bi_char[:20]

for i,j in top_twenty:
    print(i , " : " , j)


```

```python
###################SYLLABLES ##########################
# inputs  = tokens

vowels = ['अ', 'आ', 'इ', 'ई', 'उ', 'ऊ', 'ए', 'ऐ', 'ओ', 'औ','अं']

consonants = [
    'क्', 'ख्', 'ग्', 'घ्', 'ङ्', 'च्', 'छ्', 'ज्', 'झ्', 'ञ्',
    'ट्', 'ठ्', 'ड्', 'ढ्', 'ण्', 'त्', 'थ्', 'द्', 'ध्', 'न्',
    'प्', 'फ्', 'ब्', 'भ्', 'म्', 'य्', 'र्', 'ल्', 'व्', 'श्',
    'ष्', 'स्', 'ह्'
]


v_to_m = {
    'अ' : '',
    'आ': '\u093E',  # आ
    'इ': '\u093F',  # इ
    'ई': '\u0940',  # ई
    'उ': '\u0941',  # उ
    'ऊ': '\u0942',  # ऊ
    'ए': '\u0947',  # ए
    'ऐ': '\u0948',  # ऐ
    'ओ': '\u094B',  # ओ
    'औ': '\u094C',  # औ
    'अं': 'ं',
    'अः': 'ः'
}


def Syllables(inputs):
    u=[]
    syllable=[]
    result = ''
    i=0
    while i < len(inputs):
        if inputs[i] in vowels:  
            result = result + inputs[i]
            if(i+1 < len(inputs) and inputs[i+1]=='अं'):
                result = result + inputs[i+1]
                i=i+1
            u.append(result)
            result=""

        else:
            result += inputs[i]
        i+=1



    # print(u)
    for i in u:
        str=""
        for j in i:
            if j in vowels:
                if len(str)==0:
                    str = str + j
                    continue
                elif(len(str)>0):
                    h = str[len(str)-1]
                    if (h==halant):
                        str = str[:-1]
                        str=str + (v_to_m[j])
                    else:
                        str = str + v_to_m[j]

                else :
                    str=str + j
            else:
                str=str+j        
        syllable.append(str) 
    return syllable
```
